```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import pandas

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

fathi = webdriver.Chrome()

fathi.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")

wait = WebDriverWait(fathi, 15)

# wait for table rows to load
rows = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//table//tr")))

movies = []

# skip header row
for row in rows[1:]:
    try:
        rank = row.find_element(By.XPATH, "./td[1]").text
        title = row.find_element(By.XPATH, "./td[2]").text
        gross = row.find_element(By.XPATH, "./td[3]").text
        year = row.find_element(By.XPATH, "./td[4]").text

        print(rank, title, "=>", gross, "|", year)

        movies.append([rank, title, gross, year])

    except:
        continue

# save to CSV
df = pandas.DataFrame(movies, columns=["Rank", "Title", "Lifetime Gross", "Year"])
df.to_csv("top_200_movies.csv", index=False)

fathi.quit()
```
