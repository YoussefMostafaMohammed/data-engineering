```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import pandas

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
fathi=webdriver.Chrome()
fathi.get("https://www.amazon.com/gp/goldbox?ref_=nav_cs_gb")
wait = WebDriverWait(fathi, 15)

cards = fathi.find_elements(By.XPATH, "//div[@data-testid='product-card']")

for card in cards:
    title = card.find_element(By.XPATH, ".//span[@class='a-truncate-cut']").text    
    price = card.find_element(By.XPATH, ".//span[@class='a-price-whole']").text
    print(title, "=>", price)
```
