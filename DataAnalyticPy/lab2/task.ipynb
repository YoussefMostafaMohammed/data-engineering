### Task 2: Deep Cleaning (The "Janitor" Phase)
Write the code to perform the following cleaning steps:

1. Duplicates: Remove any duplicate rows from the dataset.

````python
df = df.drop_duplicates()
print(df)
````

2. Date Parsing: Convert the Date column to datetime objects. Coerce any errors (like "ERROR") into NaT, then drop the rows with missing dates.

````python

df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df = df.dropna(subset=['Date'])
print(df)
df.info()
df.to_csv("cleaned_data.csv")
````

3. String Cleaning: The Sales_Amount column has symbols ($ and USD). Remove them and convert the column to a float.

````python
df['Sales_Amount'] = (df['Sales_Amount'].str.replace("$","",regex=False))
df['Sales_Amount'] = (df['Sales_Amount'].str.replace("USD","",regex=False))
df['Sales_Amount'] = df['Sales_Amount'].astype(float) 
df.to_csv("cleaned_data.csv")
df.info()
````

4. Imputation: Fill the missing (NaN) values in the Region column with the word "Unknown".

````python
df['Region'] = df['Region'].fillna("Unknown")
df.to_csv("cleaned_data.csv")
df.info()
print(df['Region'].isna().sum())
````
