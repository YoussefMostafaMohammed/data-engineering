Exercise 4.2: The Complex Query
Setup: ages = np.random.randint(18, 90, 100) and income = np.random.randint(20000, 150000, 100).
Task:

```python

ages = np.random.randint(18, 90, 100)
income = np.random.randint(20000, 150000, 100)
```

1. Find the average income of people older than 50.

```python
mask_over_50 = ages > 50
average_income_over_50 = income[mask_over_50].mean()
print("Average income of people older than 50:", average_income_over_50)
```

2. Find the count of people who are younger than 30 AND have an income greater than 100,000.

```python
mask_young_and_rich = (ages < 30) & (income > 100000)
print(mask_young_and_rich)
counter = np.sum(mask_young_and_rich)
print("Count of people younger than 30 with income > 100,000:", counter)
```
